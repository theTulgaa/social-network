```python
import networkx as nx

def read_net_w(filename):
    g = nx.Graph()
    with open(filename) as f:
        f.readline()  # Skip the header
        for l in f:
            l = l.strip().split(",")  # Remove extra spaces/newlines and split by comma
            g.add_edge(l[0], l[1], weight=int(l[2]))  # Use the correct list indexing
    return g


g = read_net_w('/content/drive/MyDrive/niigem/asoiaf-book3-edges.csv')
```
